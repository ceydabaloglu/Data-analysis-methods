**Verilerin okunması yapıldı**

```python
import pandas as pd
import numpy as np

path ="strdata.txt"
data = pd.read_csv(path,sep="\s+")

data
```

1. Fassen Sie Namen und Nachnamen in einer einzigen Spalte zusammen.


```python
data["FullName"] = data.Name.astype(str) + data.Surname.astype(str)
data
```

2. Konvertieren Sie "Land" in einen kategorischen Datentyp und kodieren Sie dann die Länder als Ganzzahlen (z. B. Telefonvorwahlen).

```python
data["Country_code"] = data['Country'].astype("category").array.codes

data
data["Country_code"]=data["Country_code"].replace([0,1,2],[+49,+90,+44])

data

```

3. Zerlegen Sie die Telefonnummern in Ortsvorwahl und Rufnummer.

```python
arr1 =[] 
arr2 =[]

x=0
for i in data.Tel:
   arr1.append(str(i)[:3])
   arr2.append(str(i)[3:9])

data["area_code"] = arr1
data["tel_no"] =arr2

data
```

4. Splitting (using regexp) usernames, domain names, and suffixes into separate columns.

```python
import re

arr0 = []
arr1 = []
arr2 = []

pattern = r"([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})"
regex = re.compile(pattern, flags=re.IGNORECASE)
for i in data["e-mail"]:
    reg = regex.match(str(i))
    arr0.append(reg.groups()[0])
    arr1.append(reg.groups()[1])
    arr2.append(reg.groups()[2])

data['benutzername'] = arr0
data['domain'] = arr1
data['suffix'] = arr2

data
```

5. Parse the computer language names (use regexp) into separate columns and encode them.
