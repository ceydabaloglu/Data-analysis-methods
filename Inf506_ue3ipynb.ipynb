```python
import pandas as pd
import json
import requests

url = ("https://raw.githubusercontent.com/wesm/pydata-book/3rd-edition/datasets/usda_food/database.json")
res =requests.get(url)
res.raise_for_status()
res



```

```python
db = res.json()
print(db)

print(len(db))
```

```python
db[0].keys()
```

```python
db[0]["nutrients"][0]
```

```python
info_keys =["description" , "group", "id", "manufacturer"]
info = pd.DataFrame(db, columns=info_keys)

info.head()
```

```python
info.info()
#info.info() çıktısından üretici sütununda eksik veri olduğunu görebiliriz.
```

```python
pd.value_counts(info["group"])[:10]
```

```python
nutrients = []

for rec in db:
  fnuts = pd.DataFrame(rec["nutrients"])
  fnuts["id"] = rec["id"]
  nutrients.append(fnuts)

nutrients = pd.concat(nutrients,ignore_index=True)
#Her besin listesi DataFrame dönüştürülür
#Boş bir yiyecek sütunu açılır.
#dataframe concat koutu ile bu listeye eklenir

nutrients
```

```python
nutrients.duplicated().sum()

nutrients = nutrients.drop_duplicates()
nutrients
```

```python
from numpy import False_
col_mapping = {"description" : "food",
               "group":"fgroup"}

info = info.rename(columns=col_mapping , copy=False)
info.info()
```

```python
col_mapping = {"description" : "food",
               "group":"fgroup"}
              
nutrients = nutrients.rename(columns=col_mapping , copy=False)
nutrients
```

```python
ndata = pd.merge(nutrients, info ,on="id")

print(ndata.info())
print("--------------------------------------------------------")
ndata.iloc[30000]

```

```python
result = ndata.groupby(["nutrient","fgroup"])["value"].quantile(0.5)

result["Zinc, Zn"].sort_values().plot(kind="barh")
```
