```python
import seaborn as sns
import joblib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


```

```python
df = joblib.load('dfo.joblib')
df
```

```python
fig, axes = plt.subplots(2, 3, figsize=(10,8))

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)

axes[0,0].scatter (df.lon,df.lat,alpha=0.2 , s=2 ,color ="green" )
axes[0,1].scatter (df.year,df.lon,alpha=0.2 , s=2,color ="blue")
axes[0,2].scatter (df.year,df.lat,alpha=0.2 , s=2,color ="pink")
axes[1,2].scatter (df.year,df.carea,alpha=0.2 , s=2, color ="yellow")
axes[1,1].hist (df.carea,20,log=True ,alpha=0.2, color= "black")
axes[1,0].hist (df.lat,20,alpha=0.2 )

```

```python
axes[0,0].set_xlabel("Longitude") 
axes[0,0].set_ylabel("Latitude")

axes[0,1].set_xlabel("Year") 
axes[0,1].set_ylabel("Longitude")

axes[0,2].set_xlabel("Year") 
axes[0,2].set_ylabel("Latitude")

axes[1,2].set_xlabel("Year") 
axes[1,2].set_ylabel("Carea")

axes[1,2].set_xlabel("Year") 
axes[1,2].set_ylabel("Carea")

axes[1,1].set_xlabel("Carea") 
axes[1,1].set_ylabel("Frequency")

axes[1,0].set_xlabel("Latitude") 
axes[1,0].set_ylabel("Frequency")

fig
```
